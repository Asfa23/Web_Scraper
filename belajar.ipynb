### ( **Mengambil elemen yang dipilih** )

Membuat variabel baru untuk mengambil div tertentu dengan nama class tertentu dan menggunakan bantuan for loop agar dapat berulang. Didalam for loop tersebut terdapat tambahan find untuk mencari lebih spesifik apa yang akan dicari.


```python
#dibikin data frame
from sqlite3 import Row
import pandas as pd

nama_elemen = [] #membuat variabel kosong

elemen = soup.find_all("div", class_="elementor-button-wrapper") 

for x in elemen: #menggunakan bantuan for loop agar dapat diambil semuanya
    nama = x.find('span','elementor-button-text').get_text() 
    nama_elemen.append(nama) #menambahkannya pada nama variebel kosong 'nama_elemen'

dict = {'Judul': nama_elemen} #membuat dictionary
df = pd.DataFrame(dict, columns = ["Judul"]) #membuat data framenya
df.to_csv('Scrapping_Hebat_News.csv')

df
```

### ( **Membuat data frame** )

Diperlukan library tambahan bernama pandas untuk membuat data frame. Kita perlu membuat list kosong yang nantinya akan diisi perulangan. Lakukan for loop dengan tambahan append didalamnya agar list kosong yang sudah kita buat tadi, dapat terisi.'.get_text()' digunakan untuk mengambil teksnya saja. Kurang lebih untuk membuat dictionary terdapat pada kodingan di atas. Disitu saya membuat title dengan nama 'judul' dan hasil data frame tersebut di convert menjadi .csv.

```python
#crawling

import csv
import requests

list_kosong1 = [] #list kosong untuk diisi judul
list_kosong2 = [] #list kosong untuk diisi tanggal

elemen_berita = soup.find_all('div', class_='elementor-post__card') #mencari elemen tertentu

for y in range(1,3): #pengulangan untuk beberapa page
    link = "https://www.unair.ac.id/category/berita/page/{}/".format(y) #link awal untuk diakses
    req  = requests.get(link) #scrapping menggunakan request get
    soup2 = BeautifulSoup(req.text,'html.parser') #scrapping menggunakan beautifulsoup
 
    for z in elemen_berita: #pengulangan untuk mengambil elemennya
        title = z.find('h3', class_='elementor-post__title').get_text().strip() 
        date  = z.find('span', class_='elementor-post-date').get_text().strip()
        list_kosong1.append(title) #memasukkan text yang diambil kedalam list_kosong1
        list_kosong2.append(date) #memasukkan text yang diambil kedalam list_kosong2

dict2 = {'title': list_kosong1, 'date': list_kosong2} #membuat dictionary
df2 = pd.DataFrame(dict2, columns = ["title","date"]) #membuat data framenya
df2.to_csv('Crawling_Hebat_News.csv')

df2
```
